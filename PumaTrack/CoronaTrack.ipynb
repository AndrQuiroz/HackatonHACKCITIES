```
import gpxpy 
import gpxpy.gpx 
import pandas as pd

gpx_file = open('3257507.gpx', 'r') 
gpx_file2 = open('3257505.gpx', 'r') 
gpx = gpxpy.parse(gpx_file) 
gpx2 = gpxpy.parse(gpx_file2) 
```

Importamos archivos gpx generados por co-track.ia importar gpx //instalar el pip

```
lat1 = open("lat1.txt","w")
long1 = open("long1.txt","w")
lat2 = open("lat2.txt","w")
long2 = open("long2.txt","w")
for track in gpx.tracks: 
  for segment in track.segments:  
    for point in segment.points: 
      
      lat1.write(str(point.latitude)+"\n")
      long1.write(str(point.longitude)+"\n")
lat1.close()
long1.close()
for track in gpx2.tracks: 
  for segment in track.segments:  
    for point in segment.points: 
      lat2.write(str(point.latitude)+"\n")
      long2.write(str(point.longitude)+"\n")
lat2.close()
long2.close()
```

```
def dist(lon1, lat1, lon2, lat2): 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    km = 6371* c
    return km*100


if(lati1.shape[0]<lati2.shape[0]):
  for i in range(4422):
    a = long1.values[i]
    b = lati1.values[i]
    c = long2.values[i]
    d = lati2.values[i]
    if (dist(a,b,c,d)<20):
      print("EMERGENCIA ALERTA DE CONTAGIO")
      print(lati1.values[i],long1.values[i])
      print(lati2.values[i],long2.values[i])


```

```
from datetime import datetime, timedelta




def hora(hora_cad1,hora_cad2):
  hora1 = datetime.strptime(hora_cad1, '%d-%m-%Y %H:%M')
  hora2 = datetime.strptime(hora_cad2, '%d-%m-%Y %H:%M')
  res = hora2 - hora1
  return res



print(hora('01-03-2019 18:00','01-03-2019 18:04'))
```
